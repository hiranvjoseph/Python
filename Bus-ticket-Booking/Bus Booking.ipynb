```python
import tkinter as tk

class BusTicketBookingApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Bus Ticket Booking App")

        self.seat_count = 40
        self.available_seats = list(range(1, self.seat_count + 1))
        self.booked_seats = []

        self.initialize_ui()

    def initialize_ui(self):
        self.label = tk.Label(self.root, text="Bus Ticket Booking")
        self.label.pack()

        self.label2 = tk.Label(self.root, text="Available Seats: " + str(self.seat_count))
        self.label2.pack()

        self.reserve_button = tk.Button(self.root, text="Reserve Seat", command=self.reserve_seat)
        self.reserve_button.pack()

        self.cancel_button = tk.Button(self.root, text="Cancel Reservation", command=self.cancel_reservation)
        self.cancel_button.pack()

        self.quit_button = tk.Button(self.root, text="Quit", command=self.root.destroy)
        self.quit_button.pack()

    def reserve_seat(self):
        if not self.available_seats:
            tk.messagebox.showinfo("No Seats", "Sorry, no seats available.")
        else:
            seat_number = self.available_seats.pop(0)
            self.booked_seats.append(seat_number)
            self.label2.config(text="Available Seats: " + str(len(self.available_seats)))
            tk.messagebox.showinfo("Seat Reserved", f"Seat {seat_number} reserved.")

    def cancel_reservation(self):
        if not self.booked_seats:
            tk.messagebox.showinfo("No Reservations", "You have no reservations to cancel.")
        else:
            seat_number = self.booked_seats.pop()
            self.available_seats.append(seat_number)
            self.available_seats.sort()
            self.label2.config(text="Available Seats: " + str(len(self.available_seats)))
            tk.messagebox.showinfo("Reservation Canceled", f"Reservation for seat {seat_number} canceled.")

if __name__ == "__main__":
    root = tk.Tk()
    app = BusTicketBookingApp(root)
    root.mainloop()
```

```python

import tkinter as tk
from tkinter import messagebox 
class BusTicketBookingApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Bus Ticket Booking App")

        self.trip_routes = {
            "Route 1": {
                "available_seats": list(range(1, 21)),
                "booked_seats": [],
            },
            "Route 2": {
                "available_seats": list(range(1, 21)),
                "booked_seats": [],
            },
            "Route 3": {
                "available_seats": list(range(1, 21)),
                "booked_seats": [],
            },
            "Route 4": {
                "available_seats": list(range(1, 21)),
                "booked_seats": [],
            },
            "Route 5": {
                "available_seats": list(range(1, 21)),
                "booked_seats": [],
            }
        }

        self.user_data = {}

        self.initialize_ui()

    def initialize_ui(self):
        self.label = tk.Label(self.root, text="Bus Ticket Booking")
        self.label.pack()

        self.user_id_label = tk.Label(self.root, text="User ID:")
        self.user_id_label.pack()

        self.user_id_entry = tk.Entry(self.root)
        self.user_id_entry.pack()

        self.user_name_label = tk.Label(self.root, text="User Name:")
        self.user_name_label.pack()

        self.user_name_entry = tk.Entry(self.root)
        self.user_name_entry.pack()

        self.departure_label = tk.Label(self.root, text="Departure:")
        self.departure_label.pack()

        self.departure_entry = tk.Entry(self.root)
        self.departure_entry.pack()

        self.destination_label = tk.Label(self.root, text="Destination:")
        self.destination_label.pack()

        self.destination_entry = tk.Entry(self.root)
        self.destination_entry.pack()

        self.route_label = tk.Label(self.root, text="Select Trip Route:")
        self.route_label.pack()

        self.route_var = tk.StringVar(self.root)
        self.route_var.set("Route 1")  # Default route

        self.route_menu = tk.OptionMenu(self.root, self.route_var, *self.trip_routes.keys())
        self.route_menu.pack()
        
        self.trip_status_label = tk.Label(self.root, text="Trip Status:")
        self.trip_status_label.pack()

        self.trip_status_text = tk.Text(self.root, height=10, width=40)
        self.trip_status_text.pack()

        self.reserve_button = tk.Button(self.root, text="Reserve Seat", command=self.reserve_seat)
        self.reserve_button.pack()

        self.cancel_button = tk.Button(self.root, text="Cancel Reservation", command=self.cancel_reservation)
        self.cancel_button.pack()

        self.quit_button = tk.Button(self.root, text="Quit", command=self.root.destroy)
        self.quit_button.pack()

    def reserve_seat(self):
        user_id = self.user_id_entry.get()
        user_name = self.user_name_entry.get()
        route = self.route_var.get()

        if user_id not in self.user_data:
            self.user_data[user_id] = {
                "user_name": user_name,
                "reservations": {},
            }

        if not self.trip_routes[route]["available_seats"]:
            messagebox.showinfo("No Seats", "Sorry, no seats available for this route.")
        else:
            seat_number = self.trip_routes[route]["available_seats"].pop(0)
            self.trip_routes[route]["booked_seats"].append(seat_number)
            self.user_data[user_id]["reservations"][route] = seat_number
            message = f"Seat {seat_number} reserved for {user_name} on {route}."
            messagebox.showinfo("Seat Reserved", message)

            # Append the reservation message to the trip status text widget
            self.trip_status_text.insert(tk.END, message + "\n")
    def update_trip_status(self):
        self.trip_status_text.delete(1.0, tk.END)  # Clear the existing text

        for route, data in self.trip_routes.items():
            available_seats = len(data["available_seats"])
            booked_seats = len(data["booked_seats"])
            status = f"{route}: Available seats - {available_seats}, Booked seats - {booked_seats}\n"
            self.trip_status_text.insert(tk.END, status)
            self.update_trip_status()
    
    def cancel_reservation(self):
        user_id = self.user_id_entry.get()

        if user_id not in self.user_data:
            messagebox.showinfo("No Reservations", "You have no reservations to cancel.")
            return

        route = self.route_var.get()

        if route in self.user_data[user_id]["reservations"]:
            seat_number = self.user_data[user_id]["reservations"][route]
            self.trip_routes[route]["booked_seats"].remove(seat_number)
            self.trip_routes[route]["available_seats"].append(seat_number)
            del self.user_data[user_id]["reservations"][route]
            messagebox.showinfo("Reservation Canceled", f"Reservation for seat {seat_number} on {route} canceled.")
            self.update_trip_status()
            
if __name__ == "__main__":
    root = tk.Tk()
    app = BusTicketBookingApp(root)
    root.mainloop()
```
