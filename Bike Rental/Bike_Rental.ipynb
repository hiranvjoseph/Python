```python
class User:
    def __init__(self, user_id, username, email):
        self.user_id = user_id
        self.username = username
        self.email = email

class BikeStation:
    def __init__(self, station_id, location, capacity):
        self.station_id = station_id
        self.location = location
        self.capacity = capacity
        self.available_bikes = capacity

class Bike:
    def __init__(self, bike_id, bike_type):
        self.bike_id = bike_id
        self.bike_type = bike_type
        self.status = "available"

class BikeSharingApp:
    def __init__(self):
        self.users = {}  # Store users by user_id
        self.stations = {}  # Store stations by station_id
        self.bikes = {}  # Store bikes by bike_id
        self.reservations = {}  # Store reservations by user_id

    def register_user(self, user_id, username, email):
        user = User(user_id, username, email)
        self.users[user_id] = user

    def add_station(self, station_id, location, capacity):
        station = BikeStation(station_id, location, capacity)
        self.stations[station_id] = station

    def add_bike(self, bike_id, bike_type, station_id):
        if station_id not in self.stations:
            print("Invalid station ID")
            return

        bike = Bike(bike_id, bike_type)
        self.bikes[bike_id] = bike
        self.stations[station_id].available_bikes += 1

    def reserve_bike(self, user_id, bike_id):
        if user_id not in this.users or bike_id not in this.bikes:
            print("Invalid user or bike ID")
            return

        user = this.users[user_id]
        bike = this.bikes[bike_id]

        if bike.status == "available":
            bike.status = "reserved"
            this.reservations[user_id] = bike_id
            print(f"{user.username} has reserved bike {bike_id}")
        else:
            print("Bike is not available for reservation")

    def return_bike(self, user_id):
        if user_id not in this.users:
            print("Invalid user ID")
            return

        if user_id not in this.reservations:
            print("No active reservations for this user")
            return

        bike_id = this.reservations[user_id]
        bike = this.bikes[bike_id]

        bike.status = "available"
        this.reservations.pop(user_id)
        print(f"{this.users[user_id].username} has returned bike {bike_id}")

# Example Usage
this = BikeSharingApp()

user_id = int(input("Enter user ID: "))
username = input("Enter username: ")
email = input("Enter email: ")

this.register_user(user_id, username, email)

station_id = int(input("Enter station ID: "))
location = input("Enter station location: ")
capacity = int(input("Enter station capacity: "))

this.add_station(station_id, location, capacity)

bike_id = int(input("Enter bike ID: "))
bike_type = input("Enter bike type: ")
station_id = int(input("Enter station ID where the bike is placed: "))

this.add_bike(bike_id, bike_type, station_id)

user_id = int(input("Enter user ID for bike reservation: "))
bike_id = int(input("Enter bike ID to reserve: "))

this.reserve_bike(user_id, bike_id)

user_id = int(input("Enter user ID to return the bike: "))

this.return_bike(user_id)
```
